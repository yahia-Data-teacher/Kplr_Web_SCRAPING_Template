```python
%cd /content/drive/MyDrive/vin_decoder/
nb=1460
n={'Adress line 1':'Adress_line_1', 'Adress line 2':'Adress_line_2',
       'Model year':'Model_year','Trim level':'Trim_level', 'Body style':'Body_style', 'Engine type':'Engine_type',
                   'Fuel type':'Fuel_type','Vehicle class':'Vehicle_class', 'Vehicle type':'Vehicle_type',
                   'Manufactured in':'Manufactured_in', 'Assembly plant':'Assembly_plant'}
import pandas as pd
df_vin=pd.read_csv('vin_11_temp.csv')
from tqdm.notebook import tqdm
import json
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
vin=list()
for i in tqdm(df_vin['vin_11'][nb:nb+20], desc='First Octet'):#df_vin['vin_11'][nb:nb+20] vin_ok
  d={}
  d={'vin_12':i,
      'vin_11':i[:11],
      'vin_pref':i[:8]+i[9]
      }
  quotes_page='https://vindecoder.pl/'+i+'692889'
  uClient=uReq(quotes_page)
  page_html = uClient.read()
  uClient.close()
  page_soup = soup(page_html, 'html.parser')
  tab= page_soup.findAll('table' ,attrs={'class':"table table-striped table-two-col"})
  for quote in tqdm(tab[:3:2], desc='Second Octet',leave=False):
    for j in tqdm(quote, desc='Third Octet',leave=False):
      if str(j.find('th').text.strip()) in n.keys():
        d[n.get(j.find('th').text.strip())]=j.find('td').text.strip()
      else:
        d[j.find('th').text.strip()]=j.find('td').text.strip()
  vin.append(d)
print(d)
with open('vin_11_temp.json', 'a') as outfile:
    json.dump(vin, outfile)

print(df_vin['vin_11'][nb+19])
print(str(nb)+' : '+str(nb+20))
```

```python
#Getting all Job Links
links = []
No_Pages = int(input("How many pages you want to scrape: "))
try:
    for page in range(2,No_Pages+3):
        time.sleep(5)
        jobs_window = driver.find_element(By.CLASS_NAME,'jobs-search-results-list')
        jobs_list = jobs_window.find_elements(By.CSS_SELECTOR,".jobs-search-results__list-item")
        for job in jobs_list:
            all_links = job.find_elements(By.TAG_NAME,'a')
            for a in all_links:
                if str(a.get_attribute('href')).startswith("https://www.linkedin.com/jobs/view") and a.get_attribute('href') not in links: 
                    links.append(a.get_attribute('href'))
                else:
                    pass
                driver.execute_script("arguments[0].scrollIntoView();", job)
        page=page-1
        driver.find_element(By.XPATH,f"//button[@aria-label='Page {page}']").click()
        time.sleep(5)
except:
    pass
print('Found ' + str(len(links)) + ' links for job offers')
```
